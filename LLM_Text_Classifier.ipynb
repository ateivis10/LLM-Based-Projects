**Afterward, we use some pandas magic to create a dataframe.**

```python
import pandas as pd

class_list = ['INSULT', 'ABUSE', 'PROFANITY', 'OTHER']

# Load the uploaded files into pandas DataFrames
df1 = pd.read_csv('germeval2019GoldLabelsSubtask1_2.txt', sep='\t', lineterminator='\n', encoding='utf8', names=["tweet", "task1", "task2"])
df2 = pd.read_csv('germeval2019_Testdata_Subtask12.txt', sep='\t', lineterminator='\n', encoding='utf8', names=["tweet", "task1", "task2"])

# Check the number of rows and columns
print("Dataset Shape:", df1.shape)
print("Dataset Shape:", df2.shape)
# Display the number of rows and columns
print(f"The dataset contains {df1.shape[0]} rows and {df1.shape[1]} columns.")
print(f"The dataset contains {df2.shape[0]} rows and {df2.shape[1]} columns.")

# Concatenate the two DataFrames
df = pd.concat([df1, df2])

# Clean the 'task2' column by removing any carriage return characters
df['task2'] = df['task2'].str.replace('\r', "")

# Fill missing values in 'task2' with a placeholder (e.g., 'OTHER')
df['task2'].fillna('OTHER', inplace=True)  # Replace NaN with 'OTHER'

# Create a new column 'pred_class' that maps 'task2' to the class list index
df['pred_class'] = df.apply(lambda x: class_list.index(x['task2']), axis=1)

# Keep only the 'tweet' and 'pred_class' columns
df = df[['tweet', 'pred_class']]

# Print the shape and the first few rows of the DataFrame
print(df.shape)
df.head()
```

**Since we don't have a test dataset, we split our dataset — train_df and test_df. We use 90% of the data for training (train_df) and 10% for testing (test_df).**

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df, test_size=0.10)

print('train shape: ',train_df.shape)
print('test shape: ',test_df.shape)

# train shape:  (6309, 2)
# test shape:  (702, 2)
```

**Saving the Model**

```python
import os
import tarfile

def pack_model(model_path='/content/drive/MyDrive', file_name=''):
  # Check if the directory exists before attempting to walk through it
  if os.path.exists(model_path):
    # Get the list of files (handling potential empty directories)
    files_list = [files for root, dirs, files in os.walk(model_path)]
    # Check if any files were found before proceeding
    if files_list and files_list[0]:  # Check if files_list is not empty and its first element is not empty
      files = files_list[0]
      with tarfile.open(file_name + '.tar.gz', 'w:gz') as f:
        for file in files:
          f.add(f'{model_path}/{file}')
    else:
      print(f"No files found in the directory: {model_path}")
  else:
    print(f"The directory does not exist: {model_path}")

# Run the function (using outputs/ as the model path since the model is likely saved there)
pack_model('outputs/', 'model_name')
```
